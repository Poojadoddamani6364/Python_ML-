# **1.NUMPY**

- **Create Array**

```python
import numpy as np
a=np.array([4,5,1,2,3])
b=np.arange(2,4)
c=np.zeros((1,4))
d=np.ones((4,5))

print("Array a:",a)
print("Array b:",b)
print("1x4 Zeros:",c)
print("4x5 Ones:",d)
```

- **Indexing and Slicing**

```python
print("First element of a:",a[0])
print("Last two elements of a:",b[-2:])
```

- **Basic Arithematic Operations**

```python
print("a+3:",a+3)
print("a*2:",a*2)
```

- **Basic statistics**

```python
print("Square of a:",np.square(a))
print("Mean of a:",np.mean(a))
print("Standard deviation of a:",np.std(a))
```

- **Reshaping Arrays**

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaping 3x3 matrix:")
matrix
```

- **Transpose**

```python
print("Transpose")
matrix.T
```


- **Boolean Indexing**

```python
print("Elements >= 4:",matrix[matrix >= 4])
```

# **2.PANDAS**

- **create dataframe manually**

```python
import pandas as pd

data={
    'name':['Alice','Bob','Charlie','David','Eve'],
    'age':[25,30,35,40,22],
    'Department':['sales','engineering','marketing','sales','engineering'],
    'Salary':[50000,80000,40000,90000,45000]
}
df=pd.DataFrame(data)
df
```

- **Basic Operations**

```python
print("First 3 Rows:")
df.head(3)
```

```python
print("column name:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describle numerical Columns:")
df.describe()
```

- **Filtering**

```python
print("People with salary >= 50000:")
df[df['Salary'] >= 50000]
```

- **Sorting**

```python
print("Sorted by Age:")
df.sort_values(by="age")
```

 - **Grouping and Aggregation**

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average salary by Department:")
grouped
```

- **Adding new Columns**

```python
df['Salary in Lakhs'] = df['Salary']/100000
df
```

# **3.MATPLOT**

- **LineChart**

```python
import matplotlib.pyplot as plt

years=[2019,2020,2021,2022]
revenue=[100,120,130,160]

plt.plot(years,revenue,marker='o',linestyle='-',color='pink')
plt.title("Company Revenue over Years")
plt.xlabel("Year")
plt.ylabel("Revenue(in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

days=['sun','mon','tue','wed','thu','fri','sat']
calories=[50,120,130,160,40,140,180]

plt.plot(days,calories,marker='<',linestyle='-',color='maroon')
plt.title("Weekly Report")
plt.xlabel("Days")
plt.ylabel("Calories Burnt")

plt.grid(False)
plt.show()
```

- **BarChart**

```python
departments = df['Department'].value_counts()
departments.plot(kind='bar',color='skyblue')
plt.title("Department Distribution")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

- **Histogram**

```python
ages=df['age']
plt.hist(ages,bins=5,color='pink',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

- **PieChart**

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

# **4.SEABORN**

- **Load Sample dataset**

```python
import seaborn as sns

tips = sns.load_dataset('tips')
tips.head()
```

```python
print("\n Info about the tips dataset:")
tips.info()

print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

- **Pairplot**

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

- **Heatmap**

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

- **Countplot**

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Counts of Records by day")
plt.show()
```

- **Boxplot**

- Interpreting the Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total bill by day")
plt.show()
```
