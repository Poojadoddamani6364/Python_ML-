-  **Data Preprocessing**

- Clean Column Names

```python
df.columns = df.columns.str.strip()
```

- Convert Entry Columns to Datetime

```python
df['Entry_Date'] = pd.to_datetime(df['Entry'],format='%b/%Y')
df['Entry_Year'] = df['Entry_Date'].dt.year
df.head()
```

```python
df.dtypes
```

- **Create Valuation Growth Metrics**

```python
df['Valuation_Growth'] = df['Valuation ($B)']/df['Entry Valuation^^ ($B)']
df.head()
```

- **Extract Primary City (before '/')**

```python
df['Primary_City'] = df['Location'].str.split('/').str[0]
df.head()
```

- **Count Numbers of Investors**

```python
df['Investor_Count'] = df['Select Investors'].str.count(',') +1
df.head()
```

# **Univariate Analysis**

- **Valuation Distribution Analysis**

```python
plt.hist(df['Valuation ($B)'],bins=20,alpha=0.7,color='teal',edgecolor='black')
plt.xlabel('Valuation ($B)')
plt.ylabel('Frequency')
plt.title('Distribution of Current Valuations')
plt.show()
```

# **Sector Analysis**

- **Sector Distribution**

```python
plt.figure(figsize=(12,8))

sector_counts = df['Sector'].value_counts()
colors = plt.cm.Set3(np.linspace(0,1,len(sector_counts)))

plt.pie(sector_counts.values,labels=sector_counts.index,autopct='%1.1f%%',startangle=90,colors=colors)
plt.title("Distribution of Unicorn by Sector",fontsize=16,fontweight='bold')
plt.axis("equal")
plt.show()
```

- **Displaying Key Sector**

```python
def display_top_k_sector(df,k):
  sector_counts = df['Sector'].value_counts()

  # Select top k sectors and group the rest as 'others'
  if k<len(sector_counts):
    top_sectors = sector_counts.head(k)
    other_sectors_count = sector_counts[k:].sum()
    top_sectors['Others'] = other_sectors_count
  else:
    top_sectors = sector_counts

  plt.figure(figsize=(12,8))
  colors = plt.cm.Set2(np.linspace(0,1,len(top_sectors)))

  plt.pie(top_sectors.values,labels=top_sectors.index,autopct='%1.1f%%',startangle=90,colors=colors)
  plt.title(f'Distribution of Top {k} Unicorn Sectors',fontsize=16,fontweight='bold')
  plt.axis('equal')
  plt.show()

#Call the function with k=10 to display 10 sectors
display_top_k_sector(df,10)
```

- **Top Selectors Bar Chart**

```python
plt.figure(figsize=(12,8))
top_sectors = sector_counts.head(10)
bars = plt.bar(range(len(top_sectors)),top_sectors.values,color='crimson')
plt.xlabel('Sector')
plt.ylabel('Number of Unicorns')
plt.title('Top 10 Sectors by Number of unicorns',fontsize=14,fontweight='bold')
plt.xticks(range(len(top_sectors)),top_sectors.index,rotation=90,ha='right')

#Add value lables on Bars
for bar,value in zip(bars,top_sectors.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height() +0.1,str(value),ha='center',va='bottom',fontsize=12)
plt.tight_layout()
plt.show()
```

**Key Insights**
  - EdTech leads with highest number of unicorns
  - E-Commerce and Saas are major contribution
  - EdTech emergence as significant sector
  - Diversification across Myltiple Sectors

# **Goegraphical Distribution**

- **Create Summary Table**

```python
city_summary = df.groupby('Primary_City').agg({
    'Valuation ($B)':['count','sem','mean'],
    'Entry_Year':['min','max']
}).round(2)

city_summary.columns = ['Count','Total_Valuation','Avg_Valuation','First_Entry','Last_Entry']
city_summary.head()
```

**Tempor Analysis**
 - Entry Year Trends

```python
yearly_entries = df['Entry_Year'].value_counts().sort_index()
bars = plt.bar(yearly_entries.index,yearly_entries.values,color='khaki')
plt.title("Number of unicores Entered by Year",fontsize=16,fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Number of Unicorns')
plt.xticks(yearly_entries.index,rotation=45)

#Add value labels
for bar,value in zip(bars,yearly_entries.values):
  plt.text(bar.get_x()+bar.get_width()/2,bar.get_height()+0.1,str(value),ha='center',va='bottom')

plt.tight_layout()
plt.show()
```

- **Cumulative Unicorns**

```python
cumulative = yearly_entries.cumsum()
plt.plot(cumulative.index,cumulative.values,color='brown',marker='o',linestyle='--')
plt.xlabel('Year')
plt.ylabel('Cumulative Number of Unicorns')
plt.title("Cumulative Unicorns Over Time",fontweight='bold')
plt.grid(True,alpha=0.3)
plt.show()
```

- **Average Valuation by Year**

```python
avg_val_by_year = df.groupby('Entry_Year')['Valuation ($B)'].mean()
plt.bar(avg_val_by_year.index,avg_val_by_year.values,color='coral')
plt.xlabel('Year')
plt.ylabel('Average Valuation ($B)')
plt.title("Average Valuation of Unicorns by Year",fontsize=16,fontweight='bold')
plt.show()
```

- **Sector Evoluation Over Time**

```python
def display_k_sector_evolution(df,k):
  # Group by year and sector, count occurences
  sector_year = df.groupby(['Entry_Year','Sector']).size().unstack(fill_value=0)

  # Select top k sectors based on overall count
  top_k_sectors_names = df['Sector'].value_counts().head(k).index.tolist()
  sector_year = sector_year[top_k_sectors_names]

  plt.figure(figsize=(15,8))
  sector_year.plot(kind='bar',alpha=0.7,ax=plt.gca())
  plt.xlabel('Year')
  plt.ylabel('Number of Unicorns')
  plt.title(f'Evolution of Top {k} Sectors over Time',fontsize=14,fontweight='bold')
  plt.yticks(range(0,sector_year.values.max()+5))
  plt.legend(title='Sector',loc='upper left',bbox_to_anchor=(1.05,1))
  plt.grid(axis='y',alpha=0.3)
  plt.tight_layout()
  plt.show()

#Call the function with k=5 to display top 5 sectors
display_k_sector_evolution(df,5)
```
